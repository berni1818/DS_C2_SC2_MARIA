a. Extraer la información del archivo.

```python
import pandas as pd
dfvacunas = pd.read_csv("/content/country_vaccinations.csv")
dfvacunas.head()
```

b. Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64.

```python
dfvacunas["date"] = pd.to_datetime(dfvacunas["date"])
dfvacunas.dtypes
```

c. Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).

```python
vacunas_compania = dfvacunas.groupby('vaccines')['daily_vaccinations'].sum().reset_index()
print(vacunas_compania)
```

d. Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
vacunas_mundo = dfvacunas['daily_vaccinations'].sum()
print("El total de vacunas aplicadas en el mundo fue de:",vacunas_mundo)
```

e. Calcular el promedio de vacunas aplicadas por país.



```python
vacunas_pais = dfvacunas.groupby('country')['daily_vaccinations'].mean()
print(vacunas_pais)
```

f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
vacunas_enero = dfvacunas[(dfvacunas["date"] == '2021-01-29')]
vacunas_enero_total = vacunas_enero['daily_vaccinations'].sum()
print("El total de vacuanas colocadas en 29/01/21 en todo el mundo fue de:", vacunas_enero_total)
```

g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.  

```python
dfvacunas["diferencias"] = dfvacunas["daily_vaccinations"] - dfvacunas["daily_vaccinations_raw"]
conDiferencias =dfvacunas.copy()
conDiferencias.head()
```

h. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.



```python
vacuna_antigua = dfvacunas["date"].max()
vacuna_reciente = dfvacunas["date"].min()
periodo = vacuna_antigua - vacuna_reciente
print("El periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua es:", periodo)
```

i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).

```python
dfvacunas["canVac"] = dfvacunas["vaccines"].str.split(",").str.len()
conCantidad = dfvacunas.copy()
conCantidad.head()
```

j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20 = dfvacunas[(dfvacunas["date"] < "2020-12-20")]
print("El número de registros de vacunas que se tiene antes del 20 de diciembre de 2020 es de:",len(antes20))
#print(antes20)
```

k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
pfizer = dfvacunas[dfvacunas["vaccines"].str.contains("Pfizer", case=False, na=False)]
print("El número de registros donde se uso la vacuna Pfizer son de:", len(pfizer))
#print(pfizer)
```

l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter.

```python
!pip install XlsxWriter
with pd.ExcelWriter("resultadosReto.xlsx", engine="xlsxwriter") as writer:
    conDiferencias.to_excel(writer, sheet_name="Diferencias", index=False)
    conCantidad.to_excel(writer, sheet_name="Cantidad", index=False)
    antes20.to_excel(writer, sheet_name="Antes_20dic", index=False)
    pfizer.to_excel(writer, sheet_name="Pfizer", index=False)
```
